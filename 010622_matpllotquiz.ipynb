```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 1, 2, 3])
ypoints = np.array([3,8,1,10])


plt.plot(xpoints, ypoints,ls=':',marker='o',color='red')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Python", "Maths", "English", "chemistry"])
y = np.array([35,25,15,25])
d=[0.2,0,0,0]

plt.pie(y, labels=x, explode=d)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["CSE", "Mech", "Civil", "ECE"])
y = np.array([35,25,25,15])
plt.pie(y, labels=x)
plt.legend(title="For Subjects")
plt.show() 
```
