```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["clad", "ITP", "COMP", "ENG","EVPR","PHY","VEDC"])
mycolors=["#4CAF50","r","g","b","c","k","m"]
y = np.array([89,69,75,77,100,92,79])

plt.pie(y, labels =x,colors= mycolors)
plt.title("marks")
plt.legend(title="subjects")
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["clad", "ITP", "COMP", "ENG","EVPR","PHY","VEDC"])
y = np.array([89,69,75,77,100,92,79])
d=[0.69,0.68,0.35,0.245,0.6969,1,0.369]

plt.pie(y, labels =x,explode=d)
plt.title("marks")
plt.legend(title="subjects")
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["clad", "ITP", "COMP", "ENG","EVPR","PHY","VEDC"])
y = np.array([89,69,75,77,100,92,79])
d=[0.69,0.68,0.35,0.245,0.6969,1,0.369]

plt.pie(y, labels =x,explode=d,shadow="True")
plt.title("marks")
plt.legend(title="subjects")
plt.show() 
```
